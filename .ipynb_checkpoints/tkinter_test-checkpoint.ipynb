### Tkinter에서 위젓들을 화면에 배치하는 방식 - 3가지

#### 1. place ( 혹은 absolute )
- 위젯의 위치를 절대 좌표로 정하는 것으로, 위도우 크기 변경에 따라 위젯들이 변경되지 않으므로, 많이 사용되지 않는다.
- 이 방식은 위젯.place() 메서드를 사용한다

#### 2. pack
- 위젯들을 부모 위젯에 모두 패킹하여 불필요한 공간을 없앤다
- 위젯.pack() 메서드를 사용한다

#### 3. gird
- 위젯들을 테이블 레이아웃에 배치하는 것으로 지정된 row, column에 위젯을 놓는다.
- 위젯.grid() 메서드를 사용한다

```python
window = Tk()

label1 = Label(window, text = '이름')
label1.grid(row = 0, column = 0)

entry1 = Entry(window)
entry1.grid(row = 0, column = 1)

btn1 = Button(window, text='ok', width = 15)
btn1.grid(row = 1, column = 1)

mainloop()
```

### Tkinter 위젯
- Button : 단순한 버튼
- Label : 텍스트 혹은 이미지 표시
- CheckButton : 체크박스
- Entry : 단순한 한 라인 텍스트 박스
- ListBox : 리스트 박스
- RadioButton : 옵션 버튼
- Message : Label과 비슷하게 텍스트 표시. Label과 달리 자동 래핑 기능이 있다
- Scale : 슬라이스 바
- Scrollbar : 스크롤바
- Text : 멀티 라인 텍스트 박스로서 일부 Rich Text 기능 제공
- Menu : 메뉴 Pane
- Menubutton : 메뉴 버튼
- Toplevel : 추가로 새 윈도우 혹은 다이얼러그를 만들경우 사용
- Frame : 컨테이너 위젯. 다른 위젯들을 그룹화할 때 사용
- Canvas : 그래프와 점들로 그림을 그릴 수 있으며, 커스텀 위젯을 만드는 데 사용

```python
from tkinter.simpledialog import *
from random import *
w = Tk()
w.title('내가 만든 프로그램 ! ')
text = Text(w, height = 30, width = 80)
text.grid()
#----------------------------------------------------------변수 설정

##----------------------------------------------------------함수 정의
#----------------------------------------------------------함수 프로그램 정보
def proInfo():
    w = Tk()
    infolabel = Label(w, text='작성자 : 신경주\n전화번호 : 010-xxxx-xxxx')
    infolabel.grid()
#----------------------------------------------------------함수 새로 만들기
def newMake():
    answer = tkinter.messagebox.askyesnocancel('새로만들기', '저장하시겠습니까?')
    if answer == True :
        save()
        text.delete('0.0', END)
    elif answer == False : text.delete('0.0', END)
    else: pass
#----------------------------------------------------------함수 저장하기
def save():
    file=asksaveasfile(parent=w,mode='w')        # tkinter.filedialog 안에 asksaveasfile 이라는 함수가 있다
    print(file)
    if file!= None:
        lines = text.get('0.0', END+'-1c')       # 0행 0열부터 끝까지 get해서 lines에 저장, '-1c'는 맞미가 빈줄 넣지 말란 의미
        file.write(lines)
        file.close()
#----------------------------------------------------------함수 열기
def openfile():
#     text.delete('0.0', lines)
    file=askopenfile(parent=w, mode='r')
    print(file)
    if file!= None:
        lines = file.read()
        text.insert('0.0', lines)
        file.close() 
#----------------------------------------------------------함수 종료하기
def exit():
    if tkinter.messagebox.askokcancel('종료' ,'종료하시겠습니까?'):
        w.destroy()
#----------------------------------------------------------함수 합구하기    
def doSum():
    global entry11, sum, nn,label33
    sum = 0
    print(entry11.get())
    nn = int(entry11.get())                                # entry에 아무것도이력할시간이없었다.
    for i in range(nn+1):
        sum += i
    aa = str(sum)
    label33['text'] = aa

def canSum():
    label33['text'] = '0'
    entry11.delete(0,END)

def sumNum():
    global entry11, nn, label33
    wS = Tk()
    label11 = Label(wS, text = '1부터 어디까지 더할까요?')
    label11.grid(row = 0, column = 0)
    entry11 = Entry(wS)
    entry11.grid(row = 0, column = 1)
    btn11 = Button(wS, text = '합 구하기', command = doSum)
    btn11.grid(row = 0, column = 2)
    btn22 = Button(wS, text = '다시 하기', command = canSum)
    btn22.grid(row = 1, column = 2)
    label22 = Label(wS, text = '1부터 더한 값')
    label22.grid(row = 1, column = 0)
    label33 = Label(wS, text = '0')
    label33.grid(row = 1, column = 1)


    wS.mainloop()
#----------------------------------------------------------함수 최댓값 구하기 -1
def findLarge1():
    num = tkinter.simpledialog.askstring("최댓값 구하기","숫자를 입력하시오(1,5,66,2,..)").split(',')
    print(num)
    nums = [int(i) for i in num]
    large = nums[0]
    for i in range(1, len(nums)):
        if nums[i] > large:
            large = nums[i]
    tkinter.messagebox.askokcancel('최댓값 구하기', f'최댓값은 : {large}')
    w.destroy()    
#----------------------------------------------------------함수 최댓값 구하기 -2
def findLarge2():
    nums = [1,2,5,6,7,8,33,4,30]
    large = nums[0]
    for i in range(1, len(nums)):
        if nums[i] > large:
            large = nums[i]
    tkinter.messagebox.askokcancel('최댓값 구하기', f'최댓값은 {large}')
#----------------------------------------------------------함수 동몀이인 - 1
def sameName1():
    tkinter.simpledialog.askstring('동명이인 구하기 -1', '이름을 입력하시오(tom,jerry,..)')

#     names = name.split()
#     samename = []
#     tkinter.messagebox.askokcancel('동명이인 구하기', f'동명이인 : {samename}')
#     for i in names:
#         if names.count(i) >=2:
#             samename.append(i)
#             names.remove(i)
#     messagebox.askokcancel('동명이인 구하기', 'd')
#----------------------------------------------------------함수 동몀이인 - 2
# def sameName2():
#     name = ['tom','jerry','jane','harry',' bob','tom','son','son','kane','dele']
    
#     tkinter.messagebox.askokcancel('동명이인 구하기', f'동명이인 : {samename}')
#----------------------------------------------------------함수 온도변환
def temper():
    global entry_c, entry_f
    window = Tk()
    window.title('온도 변환 프로그램') # 윈도우 이름
    window.geometry('250x150') # 윈도우 가로 * 세로 사이즈
    label_1 = Label(window, text='온도 변환 프로그램')
    label_2 = Label(window, text = '섭씨')
    label_3 = Label(window, text = '화씨')
    label_1.grid(row=0, column = 0, columnspan =2)        # label 위젯을 윈도우에 배치 , span --> 열이나 행을 합치는 것
    label_2.grid(row=1, column = 0)
    label_3.grid(row=2, column = 0)



    entry_c = Entry(window)
    entry_f = Entry(window)

    entry_c.grid(row=1, column=1)
    entry_f.grid(row=2, column=1)                      # 형식 : Entry(객체명, 속성)

    btn_1=Button(window, text='섭씨 -> 화씨', command = cTof)
    btn_2=Button(window, text='화씨 -> 섭씨', command = fToc)

    btn_1.grid(row=3, column = 0)
    btn_2.grid(row=3, column = 1)
    window.mainloop()
    
def cTof():
    global entry_f
    c = float(entry_c.get())
    f = 9.0/5.0*c+32.0
    entry_c.delete(0,END)
    entry_f.insert(0,str(f))


def fToc():
    global entry_c
    f = float(entry_f.get())
    c = 5.0/9.0*(f-32.0)
    entry_f.delete(0,END)
    entry_c.insert(0,str(c))
    
    
#----------------------------------------------------------함수 계산기    
def fac():
    global doFac, facResult, inputNum
    window = Tk()
    window.title('팩토리얼 구하기') # 윈도우 이름
    window.geometry('250x150') # 윈도우 가로 * 세로 사이즈
    label_1 = Label(window, text='팩토리얼 구하기')
    label_2 = Label(window, text = '입력 숫자')
    label_3 = Label(window, text = '결과값')
    label_1.grid(row=0, column = 0, columnspan =2)        # label 위젯을 윈도우에 배치 , span --> 열이나 행을 합치는 것
    label_2.grid(row=1, column = 0)
    label_3.grid(row=2, column = 0)



    inputNum = Entry(window)
    facResult = Entry(window)

    inputNum.grid(row=1, column=1)
    facResult.grid(row=2, column=1)                      # 형식 : Entry(객체명, 속성)

    btn_1=Button(window, text='구하기', command = doFac)
    btn_2=Button(window, text='다시 하기', command = reFac)

    btn_1.grid(row=3, column = 0)
    btn_2.grid(row=3, column = 1)
    window.mainloop()

def doFac(inputNum):
    facinputNum = int(inputNum.get())
    facsum = 0
    facsum += facinputNum
    if facinputNum == 0 or facinputNum == 1 :return 1
    return facResult.insert(0,facsum + doFac(inputNum-1))

def reFac():
    facResult.delete(0,END)
    fac()
#----------------------------------------------------------함수 계산기
def cal():
    global col_index, s
    window = Tk()
    window.title('My Calculator')
    display = Entry(window, width = 33, bg = 'pink')
    display.grid(row = 0, column = 0, columnspan = 5)
    bt_list = '7,8,9,/,C,4,5,6,*,%,1,2,3,-,**,0,,=,+,←'.split(",")
    print(bt_list)

    col_index = 0
    row_num = 1
    for button_text in bt_list:       # command =lambda x = button_text : click(x) --> button_text 가 x로 가고 x가 click()으로 간다.
        Button(window, text=button_text, width = 5, command =lambda x = button_text : click(x)).grid(row = row_num, column = col_index)
    #     btn = Button(window, text=bt_text, width = 5)
    #     btn.grid(row = row_num, column = col_index)              
        col_index += 1
        if col_index == 5:      # row 변경 해주는 code
            row_num+=1
            col_index = 0
    def click(key):
        if key == '=':
            result = eval(display.get())
            s = str(result)
            display.insert(END, '=' + s)
        elif key == 'C':
            display.delete(0, END)               # delete(시작점,끝점) --> 시작점부터 끝점까지 지운다.
        elif key == '←':
            size = len(display.get())
            display.delete(size-1,END)
        else: display.insert(END, key)
    window.mainloop()
#----------------------------------------------------------함수 선택 정렬    
def selectSort():
    global entry_1, entry_2
    window = Tk()
    window.title('선택 정렬') # 윈도우 이름
    window.geometry('250x150') # 윈도우 가로 * 세로 사이즈
    label_1 = Label(window, text='선택 정렬')
    label_2 = Label(window, text = '입력 숫자')
    label_3 = Label(window, text = '정렬된 숫자')
    label_1.grid(row=0, column = 0, columnspan =2)        # label 위젯을 윈도우에 배치 , span --> 열이나 행을 합치는 것
    label_2.grid(row=1, column = 0)
    label_3.grid(row=2, column = 0)

    entry_1 = Entry(window)
    entry_2 = Entry(window)

    entry_1.grid(row=1, column=1)
    entry_2.grid(row=2, column=1)                      # 형식 : Entry(객체명, 속성)

    btn_1=Button(window, text='정렬하기', command = noToSort)
    btn_2=Button(window, text='숫자 재입력', command = fToc)

    btn_1.grid(row=3, column = 0)
    btn_2.grid(row=3, column = 1)
    window.mainloop()
    
def noToSort():
    num = (entry_1.get()).split(',')
    nums = [int(i) for i in num]
    
    def k(nums):
        small = nums[0]
        for i in range(1,len(nums)):
            if small > nums[i]:
                small = nums[i]
        return small

    result_list = []
    def kk(nums):
        while nums:
            result = nums.pop(nums.index(k(nums)))
            result_list.append(result)
        return entry_f.insert(0, result_list)
    
    kk(nums)
    
#----------------------------------------------------------함수 숫자 맞추기        
def guessGame():
    global count
    window = Tk()
    window.title('Up & Down 숫자 맞추기')
    label = Label(window, text = '숫자를 맞춰보세요 (1~20)')
    computerNum = randint(1,20)
    count = 0
    def guessNum():
        global count
        if count == 5:
            label['text'] = '게임이 끝났어요 정답은 : ' + str(computerNum)
            return
        count += 1
        guess = int(entry.get())
        if computerNum > guess : label['text'] = '시도 : ' + str(count) + ' 컴퓨터의 숫자가 더 커요'
        elif computerNum < guess : label['text'] = '시도 : ' + str(count) + ' 컴퓨터의 숫자가 더 작아요'
        else : label['text'] = '시도 : ' + str(count) + ' 정답'

    def nothing():
        pass

    def newGame():
        global computerNum, count
        computerNum = randint(1,20)
        label['text'] = '게임을 다시 시작합니다'
        count = 0


    entry = Entry(window)
    btn_1 = Button(window, text='확인', command = guessNum)
    btn_2 = Button(window, text='새 게임', command = newGame)
    label.grid(row = 0, column = 0, columnspan=2)
    entry.grid(row = 1, column = 0, columnspan=2)
    btn_1.grid(row=2, column=0)
    btn_2.grid(row=2, column=1)


    window.mainloop()
    
#----------------------------------------------------------함수 숫자 맞추기 - computer  
def guessGameCom():
    global label1
    w = Tk()
    w.title('내 숫자를 맞춰봐!')
    def guessEnd():
        w.destroy()
    
    def halfNum():
        global label1, start, end, half
        half = (start+end) // 2    # 값을 정수로
        label1['text']=f'당신의 숫자는 {half} 입니까?'
        
    def comGuess():
        global myNumRange, start, end, hint
        myNum = (askstring('컴퓨터가 숫자를 맞춰봅니다', '범위를 입력하세요 예 : 1 ~ 1000)').split('~'))
        myNumRange = [int (i) for i in myNum]   # myNumInt=list(map(int, num_range))
        start = myNumRange[0]
        end = myNumRange[1]
        halfNum()
        
    def youbig():
        global end, start
        end = half -1
        halfNum()
    
    def mebig():
        global end, start
        start = half +1
        halfNum()
    
    def correct():
        messagebox.askokcancel('숫자맞추기', '정답')
        
    label1 = Label(w, text='당신이 생각하는 숫자를 맞추겠습니다.', width = 30, height = 5)
    label1.grid(row = 0, column = 0)
    label2 = Label(w, text='3 ')
    label2.grid(row=0, column = 1)
    btn_1 = Button(w, text = '내 숫자가 더 크다', width = 30,height = 5, command = mebig)
    btn_1.grid(row = 1, column = 0)
    btn_2 = Button(w, text = '내 숫자가더  작다', height = 5, command = youbig)
    btn_2.grid(row=1, column =1)
    btn_3 = Button(w, text = '정답', height = 5, command = correct)
    btn_3.grid(row=1, column = 2, columnspan = 5)
    btn_4 = Button(w, text = '시작', height = 5, command = comGuess)
    btn_4.grid(row=0, column = 2)
    btn_5 = Button(w, text = '종료', height = 5, command = guessEnd)
    btn_5.grid(row=0, column = 3)
    
    w.mainloop()
#----------------------------------------------------------메뉴 설정
mymenu = Menu(w)
w.config(menu = mymenu)
filemenu = Menu(mymenu)
program = Menu(mymenu)
sortmenu = Menu(mymenu)
helpmenu = Menu(mymenu)
#----------------------------------------------------------메뉴 설정
mymenu.add_cascade(label = '파일', menu = filemenu)
mymenu.add_cascade(label = '프로그램', menu = program)
mymenu.add_cascade(label = '정렬', menu = sortmenu)
mymenu.add_cascade(label = '도움말', menu = helpmenu)
#----------------------------------------------------------파일메뉴
filemenu.add_command(label = '새로만들기', command=newMake)
filemenu.add_command(label = '열기', command=openfile)
filemenu.add_command(label = '저장하기', command=save)
filemenu.add_command(label = '종료', command=exit)
#----------------------------------------------------------프로그램 메뉴
program.add_command(label = '합 구하기', command=sumNum)                    # error ---> complete
program.add_command(label = '온도 변환', command=temper)     
program.add_command(label = '최댓값 구하기 - 1', command=findLarge1)        # error ---> complete
program.add_command(label = '최댓값 구하기(숫자 입력 x) - 2', command=findLarge2)
program.add_command(label = '동명이인 구하기 - 1', command=sameName1)       # error
program.add_command(label = '동명이인 구하기 - 2')
program.add_command(label = '팩토리얼 구하기', command=fac)                 # error
program.add_command(label = '계산기', command=cal)
program.add_command(label = '숫자 맞추기 Up & Down', command=guessGame)
program.add_command(label = '숫자 맞추기 Up & Down - com', command=guessGameCom)
#----------------------------------------------------------정렬 메뉴
sortmenu.add_command(label = '선택 정렬', command=selectSort)               # error
sortmenu.add_command(label = '삽입 정렬')
sortmenu.add_command(label = '병합 정렬')
#----------------------------------------------------------도움말 메뉴
helpmenu.add_command(label = '프로그램 정보', command = proInfo)


w.mainloop()
```

```python
from tkinter import *
window = Tk()
window.title('catch if you can')
canvas=Canvas(window, bg='white', width=500, height=500)
canvas.grid(row=0, column = 0)
x = 40
canvas.create_text(x,40, text='나 잡아봐라', tags = 'catch')
speed = 30

def stop():
    global speed
    speed = 0
    
def pause():
    global a, speed
    a = False
    
def fast():
    global speed
    speed += 80
def slow():
    global speed
    speed -= 10
def start():
    global a, speed
    a = 1
    speed = 30

frame = Frame(window)
frame.grid()
catchbtn1 = Button(frame, text='0', width=5, command=stop)
catchbtn2 = Button(frame, text='||', width=5, command=pause)
catchbtn3 = Button(frame, text='빠르게', width=5, command=fast)
catchbtn4 = Button(frame, text='느리게', width=5, command=slow)
catchbtn5 = Button(frame, text='>', width=5, command=start)
catchbtn1.grid(row=1, column=0)
catchbtn2.grid(row=1, column=1)
catchbtn3.grid(row=1, column=2)
catchbtn4.grid(row=1, column=3)
catchbtn5.grid(row=1, column=4)
a = 1

while a:
    x += 30
    if speed < 0:
        speed, x = 1, 1
    canvas.move('catch', speed, 0)        # move 형식 : move('text의 tag명', x좌표 이동거리, y좌표 이동거리)   text를 이동시킨다
    canvas.after(100)                  # after 형식 : after(화면을 멈출 시간을 ms단위로 명시)    ms초 동안 화면을 멈춘다
    canvas.update()                    # update 형식 : 화면을 업데이트 해서 다시 나타내준다
    if x > 600:
        canvas.delete('catch')
        x = 40
        canvas.create_text(x,40, text='나 잡아봐라', tags = 'catch')
window.mainloop()
```
