```python
import random 
 
def drawBoard(board):    # board: 크기 10의 리스트(0번 요소 무시) 
    print ('+---+---+---+')
    print ('| '+board[7]+' | '+board[8]+' | '+board[9]+' |')
    print ('+---+---+---+')
    print ('| '+board[4]+' | '+board[5]+' | '+board[6]+' |')
    print ('+---+---+---+')
    print ('| '+board[1]+' | '+board[2]+' | '+board[3]+' |')
    print ('+---+---+---+')

def inputPlayerLetter():
    letter = input('Do you want to be X or O? ').upper()
    if letter == 'X': return ['X', 'O']
    else:             return ['O', 'X']

def makeMove(board, letter, move): board[move] = letter
def undoMove(board, move):         board[move] = ' '
def isSpaceFree(board, move):      return board[move] == ' '

def isWin(bo, move):
    for line in '789 456 123 741 852 963 753 951'.split():
        if str(move) in line and bo[int(line[0])] == bo[int(line[1])] \
            and bo[int(line[0])] == bo[int(line[2])]: return True
    return False
               
def getPlayerMove(board):
    while True:
        move = input('What is your next move? (1-9) ')
        if move in '1 2 3 4 5 6 7 8 9'.split() and isSpaceFree(board, int(move)):
            return int(move)

def getComputerMove(board, letters):
    for letter in letters[::-1]:
        for i in range(1, 10):
            if isSpaceFree(board, i):
                makeMove(board, letter, i)
                result = isWin(board, i)
                undoMove(board, i)
                if result: return i
 
    for move in random.sample([1, 3, 7, 9], 4) + [5] + \
        random.sample([2, 4, 6, 8], 4):
        if isSpaceFree(board, move): return move
```

```python
names = ['player', 'computer']
while True: 
    theBoard = [' '] * 10
    letters = inputPlayerLetter()
    isComputerTurn = random.randint(0, 1)
    print('The ' + names[isComputerTurn] + ' will go first.')
    
    for n in range(9):
        if isComputerTurn:         move = getComputerMove(theBoard, letters)            
        else: drawBoard(theBoard); move = getPlayerMove(theBoard)
            
        makeMove(theBoard, letters[isComputerTurn], move)
        if isWin(theBoard, move):
            drawBoard(theBoard)
            if isComputerTurn: print('The computer has beaten you! You lose.')
            else:              print('Hooray! You have won the game!')
            break
            
        isComputerTurn = (isComputerTurn + 1) % 2
    else: print('The game is a tie!')        
    if not input('Do you want to play again? (yes or no) ').lower().startswith('y'): break
```
